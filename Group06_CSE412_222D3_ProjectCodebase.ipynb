```python
#Creating new dataframe to fit in the KMeans cluster to analyze our dataset hidden patterns

# Lets select the Spending score, and Annual Income Columns from the Data
x = data.loc[:, ['N','P','K','temperature','ph','humidity','rainfall']].values

# let's check the shape of x
print(x.shape)

# lets convert this data into a dataframe
x_data  = pd.DataFrame(x)
x_data.head()
```

```python
# lets determine the Optimum Number of Clusters within the Dataset

plt.rcParams['figure.figsize'] = (10, 4)

wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    km.fit(x)
    wcss.append(km.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('The Elbow Method', fontsize=20)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')

optimal_clusters = 4
plt.axvline(optimal_clusters, color='green', linestyle='--')
plt.text(optimal_clusters, max(wcss), 'Optimal Number of Clusters', horizontalalignment='right', verticalalignment='bottom', color='green', fontsize=10)

plt.show()
```

```python
# lets implement the K Means algorithm to perform Clustering analysis
km = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_means = km.fit_predict(x)

# lets find out the Results
a = data['label']
y_means = pd.DataFrame(y_means)
z = pd.concat([y_means, a], axis = 1)

z = z.rename(columns = {0: 'cluster'})

# lets check the Clusters of each Crops
print("Lets check the Results After Applying the K Means Clustering Analysis \n")
print("Crops in First Cluster:", z[z['cluster'] == 0]['label'].unique())
print("---------------------------------------------------------------")
print("Crops in Second Cluster:", z[z['cluster'] == 1]['label'].unique())
print("---------------------------------------------------------------")
print("Crops in Third Cluster:", z[z['cluster'] == 2]['label'].unique())
print("---------------------------------------------------------------")
print("Crops in Forth Cluster:", z[z['cluster'] == 3]['label'].unique())
```

```python
# Hard Clustering

print("Results for Hard Clustering\n")
counts = z[z['cluster'] == 0]['label'].value_counts()
d = z.loc[z['label'].isin(counts.index[counts >= 50])]
d = d['label'].value_counts()
print("Crops in Cluster 1:", list(d.index))
print("--------------------------------------------------")
counts = z[z['cluster'] == 1]['label'].value_counts()
d = z.loc[z['label'].isin(counts.index[counts >= 50])]
d = d['label'].value_counts()
print("Crops in Cluster 2:", list(d.index))
print("--------------------------------------------------")
counts = z[z['cluster'] == 2]['label'].value_counts()
d = z.loc[z['label'].isin(counts.index[counts >= 50])]
d = d['label'].value_counts()
print("Crops in Cluster 3:", list(d.index))
print("--------------------------------------------------")
counts = z[z['cluster'] == 3]['label'].value_counts()
d = z.loc[z['label'].isin(counts.index[counts >= 50])]
d = d['label'].value_counts()
print("Crops in Cluster 4:", list(d.index))
```

```python
# lets split the Dataset for Predictive Modelling

y = data['label']
x = data.drop(['label'], axis = 1)

print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
```

```python
# lets create Training and Testing Sets for Validation of Results
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

print("The Shape of x train:", x_train.shape)
print("The Shape of x test:", x_test.shape)
print("The Shape of y train:", y_train.shape)
print("The Shape of y test:", y_test.shape)
```

```python

# Logistic Regression
log_model = LogisticRegression()
log_model.fit(x_train, y_train)
log_y_pred = log_model.predict(x_test)

# Naive Bayes Classifier
nb_model = GaussianNB()
nb_model.fit(x_train, y_train)
nb_y_pred = nb_model.predict(x_test)

# MLP Classifier with optimization
mlp_model = MLPClassifier(hidden_layer_sizes=(100, 50),  # 2 hidden layers with 100 and 50 neurons
                          learning_rate_init=0.01,
                          max_iter=1000000,
                          solver='adam')

mlp_model.fit(x_train, y_train)
mlp_y_pred = mlp_model.predict(x_test)

# Evaluate the Model Performance
plt.rcParams['figure.figsize'] = (10, 10)

# Logistic Regression Confusion Matrix
log_cm = confusion_matrix(y_test, log_y_pred)
plt.subplot(3, 1, 1)
sns.heatmap(log_cm, annot=True, cmap='Wistia')
plt.title('Confusion Matrix for Logistic Regression', fontsize=15)

# Naive Bayes Confusion Matrix
nb_cm = confusion_matrix(y_test, nb_y_pred)
plt.subplot(3, 1, 2)
sns.heatmap(nb_cm, annot=True, cmap='Wistia')
plt.title('Confusion Matrix for Naive Bayes', fontsize=15)

# MLP Classifier Confusion Matrix
mlp_cm = confusion_matrix(y_test, mlp_y_pred)
plt.subplot(3, 1, 3)
sns.heatmap(mlp_cm, annot=True, cmap='Wistia')
plt.title('Confusion Matrix for MLP Classifier', fontsize=15)

plt.tight_layout()
plt.show()

# Classification Reports
print("Classification Report for Logistic Regression:\n", classification_report(y_test, log_y_pred))
print("Classification Report for Naive Bayes:\n", classification_report(y_test, nb_y_pred))
print("Classification Report for MLP Classifier:\n", classification_report(y_test, mlp_y_pred))

# Accuracy Scores
log_accuracy = accuracy_score(y_test, log_y_pred)
nb_accuracy = accuracy_score(y_test, nb_y_pred)
mlp_accuracy = accuracy_score(y_test, mlp_y_pred)

print(f"Accuracy of Logistic Regression: {log_accuracy}")
print(f"Accuracy of Naive Bayes: {nb_accuracy}")
print(f"Accuracy of MLP Classifier: {mlp_accuracy}")

# Determine the best model
best_model = max(("Logistic Regression", log_accuracy),
                 ("Naive Bayes", nb_accuracy),
                 ("MLP Classifier", mlp_accuracy),
                 key=lambda x: x[1])

print(f"The best model is {best_model[0]} with an accuracy of {best_model[1]:.2f}")
```
